# Step 3: Exploratory Data Analysis (EDA)
This notebook visualizes numerical feature distributions and explores payment modes, discounts, and order statuses.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

df = pd.read_csv('../data/processed/cleaned_sales.csv')
plt.style.use('seaborn-v0_8-whitegrid')
```

## Distribution of Quantity Sold
Let's plot how many items are typically purchased per transaction.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Quantity', hue='Quantity', legend=False, palette='viridis')
plt.title('Order Quantity Distribution', fontweight='bold')
plt.xlabel('Quantity')
plt.ylabel('Count')
plt.show()
```

## Distribution of Unit Price
Plotting the pricing density of products.

```python
plt.figure(figsize=(8, 4))
sns.histplot(data=df, x='UnitPrice', bins=30, kde=True, color='teal')
plt.title('Unit Price Distribution (Capped)', fontweight='bold')
plt.xlabel('Unit Price ($)')
plt.ylabel('Density')
plt.show()
```

## Discount Distribution
Visualizing what discounts are frequently applied.

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Discount', hue='Discount', legend=False, palette='coolwarm')
plt.title('Applied Discount Rate Distribution', fontweight='bold')
plt.xlabel('Discount Rate')
plt.ylabel('Transactions')
plt.show()
```

## Payment Method Distribution
Which payment channels are customers using most?

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, y='PaymentMethod', hue='PaymentMethod', legend=False, palette='Set2')
plt.title('Preferred Payment Methods', fontweight='bold')
plt.xlabel('Number of Orders')
plt.ylabel('Payment Mode')
plt.show()
```
